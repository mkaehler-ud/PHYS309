```python
# PHYS309 - HW4
# Michael Kaehler

import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

# Constant value of 2.898*10^-3 m-K.
numerator = 2.898e-3

# Temperature in Kelvin for stars, order is O5,B5,A5,F5,G5,K5,M5.
T = np.array([54000,15200,8310,6700,5660,4400,3200])

# Calculation of max wavelength intensity.
lambda_max = numerator/T

# Plot results in blue and label each point.
plt.plot(T,lambda_max,marker='o',color='blue')
plt.axis([0, 55000, 0, 1e-6])
plt.xlabel('Star Temperature (K)')
plt.ylabel('Maximum Wavelength Intensity (m)')
plt.title('Star Temperature vs Max Wavelength Intensity')
plt.text(x=T[0], y=lambda_max[0], s='O5', color='blue')
plt.text(x=T[1], y=lambda_max[1], s='B5', color='blue')
plt.text(x=T[2], y=lambda_max[2], s='A5', color='blue')
plt.text(x=T[3], y=lambda_max[3], s='F5', color='blue')
plt.text(x=T[4], y=lambda_max[4], s='G5', color='blue')
plt.text(x=T[5], y=lambda_max[5], s='K5', color='blue')
plt.text(x=T[6], y=lambda_max[6], s='M5', color='blue')
plt.show()
```

A note about units: recall from PHYS 208/245 that 1 Volt = 1 Joule/Coulomb. Electrons and protons give us a fundamental unit of electric charge, or elementary charge, $|q| = e = 1.602 \times 10^{−19}$ C. We can also calculate a fundamental unit of energy by calculating the change in kinetic energy of the elementary charge $e$ accelerated through an electric potential of 1 Volt. This new unit of energy is called the electron-Volt or eV. 1 eV = $1.602 \times 10^{−19}$ J.

You have two emitters or "photocathodes"; one is made of aluminum and the other is made of gold. On <strong>one set of axes</strong>, plot the <em>maximum</em> photoelectron kinetic energy in eV that you could measure as a function of frequency $\nu$ from <strong>both photocathodes</strong>.
```python
# Define some constant values to be used in our calculations.

h = 4.1357e-15 # Planck constant (eV-s).
phi_aluminum = 4.08 # Work function of aluminum (eV).
phi_gold = 5.10 # Work function of gold (eV).

# Define a frequency range in 10^15 Hz that has 100,000 steps.
freq_nu = np.linspace(0, 4e15, 100000)

# Calculate maximum kinetic energy for both aluminum and gold, using equation from textbook.
E_aluminum = h * freq_nu - phi_aluminum
E_gold = h * freq_nu - phi_gold

# Calculate the nu frequency cutoffs, should be vertical lines I think since frequency is on x-axis.
freq_cutoff_aluminum = phi_aluminum/h # This should calculate the threshold frequency.
freq_cutoff_gold = phi_gold/h

# Plot results in blue for aluminum and red for gold and show cutof lines as vertical dashed lines.
plt.plot(freq_nu, E_aluminum, label='Aluminum',color='blue')
plt.plot(freq_nu, E_gold, label='Gold',color='red')
plt.axvline(x=freq_cutoff_aluminum, color='blue', linestyle='--')
plt.axvline(x=freq_cutoff_gold, color='red', linestyle='--')
plt.title('Frequency vs Maximum Kinetic Energy')
plt.xlabel('Frequency (10^15 Hz)')
plt.ylabel('Maximum Kinetic Energy (eV)')
plt.legend()
plt.show()
```
